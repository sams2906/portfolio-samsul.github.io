```python
# =====================================================================
# PROJECT DATA MINING TOKO SACO.PROJECT - PROGRAM SEGMENTASI PELANGGAN
# K-MEANS CLUSTERING BERBASIS MODEL RFM (MARKETPLACE SHOPEE)
# =====================================================================

import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
# from strokeorder import plot_tools # Optional library untuk estetika grafik

# 1. DEFINISI DATASET SIMULASI BERBASIS PARAMETER RIIL TOKO SACO.PROJECT
data_saco = {
    'ID_Pelanggan': [f'USR_{i:02d}' for i in range(1, 21)],
    'Recency': [19, 17, 10, 2, 1, 3, 18, 16, 11, 9, 2, 19, 15, 12, 3, 17, 8, 10, 1, 14],
    'Frequency': [1, 1, 1, 3, 4, 3, 1, 1, 2, 2, 4, 1, 1, 2, 3, 1, 2, 1, 4, 1],
    'Monetary': [143760, 174000, 173500, 510000, 680000, 495000, 150000, 165000, 315000,
                 340000, 710000, 140000, 175000, 320000, 520000, 160000, 330000, 145000, 690000, 155000]
}

# Mengubah dictionary ke dalam bentuk Dataframe Pandas
df = pd.DataFrame(data_saco)

# 2. SELEKSI ATRIBUT UNTUK ALGORITMA K-MEANS
X = df[['Recency', 'Frequency', 'Monetary']]

# 3. EKSEKUSI ALGORITMA K-MEANS (K=3)
# Menggunakan random_state=42 agar hasil konsisten setiap kali dijalankan
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
df['Cluster'] = kmeans.fit_predict(X)

# 4. MENAMPILKAN HASIL CENTROID (PUSAT KLASTER) KONDISI AKHIR
print("==========================================================")
print("     HASIL TITIK PUSAT (CENTROID) KLASTER SACO.PROJECT     ")
print("==========================================================")
centroid = df.groupby('Cluster')[['Recency', 'Frequency', 'Monetary']].mean()
print(centroid)
print("==========================================================\n")

# 5. VISUALISASI GRAFIK 3D SCATTER PLOT
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Mendefinisikan warna untuk masing-masing klaster
colors = {0: '#10B981', 1: '#34D399', 2: '#EF4444'}
labels = {0: 'Cluster 0: Loyal Shoppers', 1: 'Cluster 1: Potential Buyers', 2: 'Cluster 2: Inactive Customers'}

# Menggambar titik koordinat objek data ke dalam grafik ruang 3D
for cluster_id in range(3):
    cluster_data = df[df['Cluster'] == cluster_id]
    ax.scatter(cluster_data['Recency'], cluster_data['Frequency'], cluster_data['Monetary'],
               c=colors[cluster_id], label=labels[cluster_id], s=100, edgecolors='black', alpha=0.8)

# Mengatur label sumbu koordinat grafik
ax.set_xlabel('Recency (Hari)', fontsize=11, fontweight='bold', labelpad=10)
ax.set_ylabel('Frequency (Kali)', fontsize=11, fontweight='bold', labelpad=10)
ax.set_zlabel('Monetary (Rupiah)', fontsize=11, fontweight='bold', labelpad=20)
ax.set_title('VISUALISASI 3D CLUSTERING PELANGGAN SACO.PROJECT', fontsize=14, fontweight='bold', pad=20)

plt.legend(loc='upper left', fontsize=10)
plt.tight_layout()

# Menampilkan grafik ke layar komputer
plt.show()

# 6. EKSPOR HASIL AKHIR KE FORMAT EXCEL UNTUK DIKUMPULKAN KE DOSEN
df.to_excel("hasil_clustering_saco_project.xlsx", index=False)
print("File 'hasil_clustering_saco_project.xlsx' berhasil diekspor untuk lampiran tugas!")
```
